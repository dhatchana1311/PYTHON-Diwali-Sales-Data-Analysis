```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load CSV file
data = pd.read_csv(r"D:\PYTHON 1\Diwali Sales Data.csv", encoding='latin1')
```

```python
print(data.info())
```

```python
product_sales = data.groupby('Product_Category')['Amount'].sum().sort_values()
product_sales.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Total Sales by Product', size= 30)
plt.xlabel('Product_Category', size= 15)
plt.ylabel('Total Sales',size= 15)
plt.show()
```

```python
daily_sales = data.groupby('Product_Category')['Amount'].sum()
daily_sales.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Daily Sales Trend')
plt.xlabel('Product_Category')
plt.ylabel('Total Sales')
plt.grid()
plt.show()
```

```python
region_sales = data.groupby('Zone')['Amount'].sum()

# Pie chart
region_sales.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=140, colors=['gold', 'skyblue', 'lightgreen'])
plt.title('Sales Contribution by Region')
plt.ylabel('')  # Removes the y-axis label
plt.show()
```

```python
data['Amount'].plot(kind='hist', bins=30, color='purple', alpha=0.7, figsize=(10, 6))
plt.title('Distribution of Sales Values')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='Product_Category', y='Amount', data=data)
plt.title('Sales Distribution Across Categories')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Orders'], data['Amount'], alpha=0.6, color='green')
plt.title('Sales vs. Quantity')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x='Zone', y='Age', palette='viridis')
plt.title('Age by Zone', fontsize=10)
plt.xlabel('Zone', fontsize=10)
plt.ylabel('Age', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=data, y='State', order=data['State'].value_counts().index, palette='cool')
plt.title('Number of Orders by State', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.show()
```

```python
category_region = data.groupby(['Product_Category', 'Zone'])['Amount'].sum().unstack()

category_region.plot(kind='bar', stacked=True, colormap='tab20', figsize=(12, 6))
plt.title('Amount by Product_Category and Zone', fontsize=16)
plt.xlabel('Product_Category', fontsize=12)
plt.ylabel('Amount', fontsize=12)
plt.legend(title='Zone')
plt.show()
```
